```python
from dataclasses import dataclass
from __future__ import annotations


@dataclass
class Segment:
    start: int
    end: int

    def isin(self, other : Segment):
        return (
            other.start >= self.start and 
            other.start <= self.end and
            other.end >= self.start and
            other.end <= self.end
        )

    def overlap(self, other : Segment):
        return (
            (
                other.start >= self.start and other.start <= self.end
            )
            or 
            (
                other.end >= self.start and other.end <= self.end
            ) 
        )

        
```

```python
from lib.utils import iter_file

score = 0
for l in iter_file('../day4.txt'):
    part1, part2 = l.split(',')
    s1, e1 = part1.split('-')
    seg1 = Segment(int(s1), int(e1))

    s2, e2 = part2.split('-')
    seg2= Segment(int(s2), int(e2))

    if seg1.overlap(seg2) or seg2.overlap(seg1):
        score += 1

print(score)

```

```python
from lib.utils import iter_file

segments = []
for l in iter_file('../day4.txt'):
    part1, part2 = l.split(',')
    s1, e1 = part1.split('-')
    seg1 = Segment(int(s1), int(e1))
    segments.append(seg1)

    s2, e2 = part2.split('-')
    seg2= Segment(int(s2), int(e2))

    segments.append(seg2)
    
segments = sorted(segments, key=lambda x: x.start)
```

```python
nb_segments = len(segments)

score = 0
curr = 0
while curr < nb_segments:
    curr_seg = segments[curr]
    curr_overlap = curr
    while curr_overlap + 1 < nb_segments:
        curr_overlap += 1
        if curr_seg.overlap(segments[curr_overlap]):
            score += 1
        else:
            break
    curr += curr_overlap
print(score)
print(nb_segments)
```
